```python
import numpy as np
import matplotlib.pyplot as plt

class Perceptron:
    def __init__(self, learning_rate=0.1, num_epochs=100):
        self.learning_rate = learning_rate
        self.num_epochs = num_epochs

    def train(self, X, y):
        self.weights = np.zeros(X.shape[1])
        self.bias = 0

        for _ in range(self.num_epochs):
            for i in range(X.shape[0]):
                prediction = self.predict(X[i])
                if prediction != y[i]:
                    update = y[i] - prediction
                    self.weights += self.learning_rate * update * X[i]
                    self.bias += self.learning_rate * update

    def predict(self, x):
        activation = np.dot(self.weights, x) + self.bias
        return 1 if activation >= 0 else 0

# Generate linearly separable data
np.random.seed(0)
num_samples = 100
X_class1 = np.random.randn(num_samples, 2) + np.array([2, 2])
X_class2 = np.random.randn(num_samples, 2) + np.array([-2, -2])
X = np.vstack((X_class1, X_class2))
y = np.hstack((np.ones(num_samples), np.zeros(num_samples)))

# Train Perceptron
perceptron = Perceptron()
perceptron.train(X, y)

# Plot data and decision boundary
plt.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.Paired, marker='o')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')

# Plot the decision boundary
x1_min, x1_max = X[:, 0].min() - 1, X[:, 0].max() + 1
x2_min, x2_max = X[:, 1].min() - 1, X[:, 1].max() + 1
xx1, xx2 = np.meshgrid(np.arange(x1_min, x1_max, 0.1),
                       np.arange(x2_min, x2_max, 0.1))
Z = np.array([perceptron.predict(np.array([x1, x2])) for x1, x2 in np.c_[xx1.ravel(), xx2.ravel()]])
Z = Z.reshape(xx1.shape)
plt.contourf(xx1, xx2, Z, alpha=0.4, cmap=plt.cm.Paired)
plt.xlim(xx1.min(), xx1.max())
plt.ylim(xx2.min(), xx2.max())
plt.title('Perceptron Decision Boundary')
plt.show()
```

```python
import numpy as np

class Perceptron:
    def __init__(self, learning_rate=0.1, num_epochs=100):
        self.learning_rate = learning_rate
        self.num_epochs = num_epochs

    def train(self, X, y):
        self.weights = np.zeros(X.shape[1])
        self.bias = 0

        for _ in range(self.num_epochs):
            for i in range(X.shape[0]):
                prediction = self.predict(X[i])
                if prediction != y[i]:
                    update = y[i] - prediction
                    self.weights += self.learning_rate * update * X[i]
                    self.bias += self.learning_rate * update

    def predict(self, x):
        activation = np.dot(self.weights, x) + self.bias
        return 1 if activation >= 0 else 0

# Generate higher-dimensional data
np.random.seed(0)
num_samples = 100
num_features = 5
X = np.random.randn(num_samples, num_features)
y = np.random.randint(2, size=num_samples)

# Train Perceptron
perceptron = Perceptron()
perceptron.train(X, y)

# Output the learned coefficients (weights and bias)
print("Learned weights:", perceptron.weights)
print("Learned bias:", perceptron.bias)
```
